```python
lineage ='immune'
lin_adata = anndata.AnnData(
            adata[adata.obs["lineage"] == lineage].X,
            obs=adata[adata.obs["lineage"] == lineage].obs,
            var=adata[adata.obs["lineage"] == lineage].var,
        )
del lin_adata.var["highly_variable"]
sc.pp.highly_variable_genes(lin_adata, n_top_genes=2000, batch_key="Run")
sc.pp.pca(lin_adata, use_highly_variable=True)
sce.pp.bbknn(
            lin_adata,
            batch_key="Run",
            n_pcs=10,
            neighbors_within_batch=5,
#     trim = 25,
            set_op_mix_ratio=0.5,
        )
# sc.pp.neighbors(lin_adata)
sc.tl.umap(lin_adata, a=2, b=1.5)
sc.tl.leiden(lin_adata)
for color in ['leiden','Cell Subtype','Run','Timepoint','Treatment',gene_dict[lineage]]:
    if color == gene_dict[lineage]:
        save = 'genes'
    else:
        save = color
    sc.pl.umap(lin_adata, color = color, save=f'{lineage}_{save}.png')
sc.tl.rank_genes_groups(lin_adata, f"leiden", method="wilcoxon")
sc.pl.rank_genes_groups_dotplot(
    lin_adata,
    groupby=f"leiden",
    n_genes=int(150 / len(lin_adata.obs[f"leiden"].unique())),
    show=True,
    save = f'{lineage}.png'
)
lin_adata.obs.groupby(['Timepoint','Treatment'])['Cell Subtype'].value_counts(normalize=True)*100
```
